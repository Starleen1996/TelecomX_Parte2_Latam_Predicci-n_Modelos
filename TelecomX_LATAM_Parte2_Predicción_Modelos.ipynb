#### Visualizaci贸n de Clientes con edad >= 65 a帽os y menor a 65 a帽os vs Vigentes/No Vigentes

```python
# Evidenciamos una superioridad en la cantidad de clientes menor a 65 a帽os, en la cual se evidencia una superioridad en el retiro de personas de la compa帽ia. Clientes No Vigentes <65 : 4508 y clientes Vigentes <65 a帽os: 1393

# Definir los colores manualmente
colores = {0: 'red', 1: 'gray'}

# Crear funci贸n para graficar con colores personalizados
def graficar_vigencia(data, titulo, ax):
    # Obtener los valores 煤nicos y contar
    counts = data['cliente_vigente'].value_counts().sort_index()

    # Dibujar barras manualmente para controlar colores
    for i, (estado, cantidad) in enumerate(counts.items()):
        ax.bar(i, cantidad, color=colores[estado])
        ax.text(i, cantidad + 1, str(cantidad), ha='center', va='bottom')

    ax.set_title(titulo)
    ax.set_xticks([0, 1])
    ax.set_xticklabels(['No Vigente', 'Vigente'])
    ax.set_ylabel('Cantidad')

# Crear subgr谩ficos
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Filtrar por edad
mayores = df_clientes_LATAM[df_clientes_LATAM['cliente>=65_anos'] == 1]
menores = df_clientes_LATAM[df_clientes_LATAM['cliente>=65_anos'] == 0]

# Graficar ambos grupos
graficar_vigencia(mayores, 'Clientes >= 65 a帽os', axes[0])
graficar_vigencia(menores, 'Clientes < 65 a帽os', axes[1])

# Leyenda personalizada
import matplotlib.patches as mpatches
leyenda = [mpatches.Patch(color='red', label='No Vigente'),
           mpatches.Patch(color='gray', label='Vigente')]
fig.legend(handles=leyenda, loc='lower center', ncol=2)

plt.tight_layout(rect=[0, 0.05, 1, 1])
plt.show()
```
