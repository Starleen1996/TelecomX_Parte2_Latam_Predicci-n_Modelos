#### Visualización de Clientes con edad >= 65 años y menor a 65 años vs Vigentes/No Vigentes

```python
# Evidenciamos una superioridad en la cantidad de clientes menor a 65 años, en la cual se evidencia una superioridad en el retiro de personas de la compañia. Clientes No Vigentes <65 : 4508 y clientes Vigentes <65 años: 1393

# Definir los colores manualmente
colores = {0: 'red', 1: 'gray'}

# Crear función para graficar con colores personalizados
def graficar_vigencia(data, titulo, ax):
    # Obtener los valores únicos y contar
    counts = data['cliente_vigente'].value_counts().sort_index()

    # Dibujar barras manualmente para controlar colores
    for i, (estado, cantidad) in enumerate(counts.items()):
        ax.bar(i, cantidad, color=colores[estado])
        ax.text(i, cantidad + 1, str(cantidad), ha='center', va='bottom')

    ax.set_title(titulo)
    ax.set_xticks([0, 1])
    ax.set_xticklabels(['No Vigente', 'Vigente'])
    ax.set_ylabel('Cantidad')

# Crear subgráficos
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Filtrar por edad
mayores = df_clientes_LATAM[df_clientes_LATAM['cliente>=65_anos'] == 1]
menores = df_clientes_LATAM[df_clientes_LATAM['cliente>=65_anos'] == 0]

# Graficar ambos grupos
graficar_vigencia(mayores, 'Clientes >= 65 años', axes[0])
graficar_vigencia(menores, 'Clientes < 65 años', axes[1])

# Leyenda personalizada
import matplotlib.patches as mpatches
leyenda = [mpatches.Patch(color='red', label='No Vigente'),
           mpatches.Patch(color='gray', label='Vigente')]
fig.legend(handles=leyenda, loc='lower center', ncol=2)

plt.tight_layout(rect=[0, 0.05, 1, 1])
plt.show()
```

## Conteo de evasión por variables numéricas
En este paso, explora cómo las variables numéricas, como "total gastado" o "tiempo de contrato", se distribuyen entre los clientes que cancelaron (evasión) y los que no cancelaron.

Este análisis ayuda a entender si ciertos valores numéricos están más asociados con la evasión, proporcionando insights sobre los factores que influyen en el comportamiento de los clientes.

```python
# Esto te permitirá ver la forma en que se distribuyen las variables numéricas según el estado de vigencia del cliente.

# Variable 1: Total gastado
plt.figure(figsize=(10, 5))
sns.histplot(data=df_clientes_LATAM, x='total_pagado_cliente', hue='cliente_vigente', kde=True, palette={1: 'gray', 0: 'red'}) # sns.histplot(...) Usa seaborn para crear un histograma de la variable total_pagado_cliente (es decir, cuánto gastó cada cliente).
#hue='cliente_vigente': Se colorean las barras según si el cliente está vigente (1) o no vigente (0).
#kde=True: Agrega la curva KDE (Kernel Density Estimation), que suaviza la forma de la distribución para mostrar tendencias más claramente.
plt.title('Distribución del Total Gastado por Estado de Vigencia')
plt.xlabel('Total Gastado')
plt.ylabel('Cantidad de Clientes')
plt.legend(title='Vigente', labels=['Si', 'No'])
plt.show()

# Variable 2: Tiempo de contrato
plt.figure(figsize=(10, 5))
sns.histplot(data=df_clientes_LATAM, x='#_meses_contrato', hue='cliente_vigente', kde=True, palette={1: 'gray', 0: 'red'})
plt.title('Distribución del Tiempo de Contrato por Estado de Vigencia')
plt.xlabel('Tiempo de Contrato (meses/años)')
plt.ylabel('Cantidad de Clientes')
plt.legend(title='Vigente', labels=['Si', 'No'])
plt.show()

#Conclusiones generales (actualizadas):
# 📉 Alta tasa de cancelación temprana: Muchos clientes se retiran en los primeros meses de contrato.

# 💰 Bajo gasto = alta rotación: Clientes que gastan poco tienden a no seguir.

# ⏳ Tiempo y gasto = vigencia: Clientes con más tiempo y gasto son más propensos a permanecer.

# 🧩 Esto puede servir para:

#Identificar perfiles de riesgo (clientes nuevos y de bajo gasto).
#Diseñar estrategias de retención más dirigidas.
```
